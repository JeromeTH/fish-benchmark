```python
from fish_benchmark.data.dataset import get_dataset

dataset = get_dataset(
    "UCF101Frames", 
    "/share/j_sun/jth264/UCF101_subset"
)
```

```python
categories = dataset.categories
print(categories)
```

```python
import os
PATH = '/share/j_sun/jth264/UCF101_subset'
for TYPE in ['train', 'val', 'test']:
    root = os.path.join(PATH, TYPE)
    for category in os.listdir(root):
        class_idx = categories.index(category)
        category_path = os.path.join(root, category)
        for video_name in os.listdir(category_path):
            video_id = video_name.split('.')[0]
            print(video_id)
            #write the annotation file with one single index: class_idx
            annotation_path = os.path.join(category_path, f'{video_id}.txt')
            with open(annotation_path, 'w') as f:
                f.write(f'{class_idx}\n')

```

```python
for root, dirs, files in os.walk(PATH):
    #remove redundant directories named 'UCF101
    if os.path.basename(root) == 'UCF101':
        os.rmdir(root)
        
```
